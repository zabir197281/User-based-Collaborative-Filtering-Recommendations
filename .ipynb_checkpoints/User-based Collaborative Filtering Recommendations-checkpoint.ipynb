# Solution for Part- (d)

```python
def get_top_similar_users(user_id, correlation_matrix, n=10):
    """Retrieve the correlation series for the user and drop the user's own correlation"""
    user_correlations = correlation_matrix[user_id].drop(user_id)
    """Sort by correlation in descending order and take the top n"""
    top_users = user_correlations.sort_values(ascending=False).head(n)
    return top_users
```

```python
similar_users=get_top_similar_users(196, correlation_matrix, n=10)
```

```python
print(f"The t 10 most similar user to user {196} is given below:\n")

for i , j in similar_users.items():
    print(f"The User {i} similarity = {round(j,4)}")
```

```python
def get_recommended_movies(user_id, user_item_matrix, correlation_matrix, n=10):
    """Get the mean ratings for each user"""
    user_mean_ratings = user_item_matrix.mean(axis=1)
    """Get the list of movies the user hasn't rated yet"""
    unrated_movies = user_item_matrix.loc[user_item_matrix[user_id].isna(), :]
    """Dictionary to store the predicted ratings"""
    predictions = {}
    
    for movie_id in unrated_movies.index:
        """Predict the rating for each unrated movie"""
        predicted_rating = predict_score(user_id, movie_id, user_item_matrix, correlation_matrix)
        predictions[movie_id] = predicted_rating
    
    """Sort the predicted ratings in descending order and get the top n movies"""
    recommended_movies = sorted(predictions.items(), key=lambda x: x[1], reverse=True)[:n]
    return recommended_movies
```

# Solution for Part- (e)

```python
import pandas as pd
import numpy as np

def improved_triangle_similarity(user_ratings_1, user_ratings_2):
    """Calculate average ratings"""
    avg_rating_1 = np.mean(user_ratings_1)
    avg_rating_2 = np.mean(user_ratings_2)
    
    """Calculate sides of the triangle for each item"""
    side_1 = np.abs(user_ratings_1 - avg_rating_1)
    side_2 = np.abs(user_ratings_2 - avg_rating_2)
    side_3 = np.abs(user_ratings_1 - user_ratings_2)
    
    """Check the triangle inequalities for each item"""
    triangle_validity = (side_1 + side_2 > side_3) & (side_1 + side_3 > side_2) & (side_2 + side_3 > side_1)
    
    """Calculate similarity"""
    similarity = np.sum(triangle_validity) / len(user_ratings_1)
    
    return similarity
```

```python
user_item_matrix = df.pivot_table(index='user_id', columns='movie_id', values='rating')
```

```python
user1 = 196
user2 = 110

"""Get the ratings for the two users, ensuring NaNs are handled"""
user1_ratings = user_item_matrix.loc[user1].dropna()
user2_ratings = user_item_matrix.loc[user2].dropna()

"""Find the common movies both users have rated"""
common_movies = user1_ratings.index.intersection(user2_ratings.index)

"""Retrieve the ratings for the common movies"""
user1_common_ratings = user1_ratings[common_movies]
user2_common_ratings = user2_ratings[common_movies]
```

```python
"""Compute the similarity score between the two users"""
similarity_score = improved_triangle_similarity(user1_common_ratings, user2_common_ratings)
print(f"The similarity score between user {user1} and user {user2} is: {similarity_score}")
```
